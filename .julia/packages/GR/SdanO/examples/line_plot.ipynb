```julia
using GR
```

```julia
x = 0:pi/100:2*pi
y = sin.(x)

plot(x,y,title="sin(x)")
```

```julia
x = -2*pi:pi/50:2*pi
y1 = sin.(x);
y2 = cos.(x);

plot(x,y1,x,y2)
```

```julia
Y = [16 2 3 13;5 11 10 8;9 7 6 12;4 14 15 1]

plot(Y)
```

```julia
x = 0:pi/100:2*pi
y1 = sin.(x)
y2 = sin.(x.-0.25)
y3 = sin.(x.-0.5)

plot(x,y1,x,y2,"--",x,y3,":")
```

```julia
x = 0:pi/10:2*pi
y1 = sin.(x)
y2 = sin.(x.-0.25)
y3 = sin.(x.-0.5)

plot(x,y1,"g",x,y2,"b--o",x,y3,"c*")
```

```julia
x = 0:0.05:10
y = cos.(5 .* x)

title("2-D Line Plot")
xlabel("X")
ylabel("cos(5x)")
plot(x,y,"-")
```
